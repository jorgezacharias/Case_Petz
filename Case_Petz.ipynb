```python
#Changing colum name 
df_vendas = pd.read_csv('/content/drive/MyDrive/projeto_dados/Vendas.csv', encoding = "ISO-8859-1", sep=';')
df_vendas.rename({'ï»¿id_data' : "Data"}, axis = 1, inplace = True)
df_vendas['Data'] = pd.to_datetime(df_vendas['Data'])
df_vendas
```

```python
#Making target for 30 days
def calcular_vendas_futuras_1(df_grupo, dias=30, min_periods=30):
    indexer = pd.api.indexers.FixedForwardWindowIndexer(window_size=dias)
    return df_grupo.shift(-1).rolling(window=indexer, min_periods=min_periods).sum()
```

```python
df['Venda_30_T'] = calcular_vendas_futuras_1(df.valor_venda)
```

```python
#Making target for 60 days
def calcular_vendas_futuras_2(df_grupo, dias=60, min_periods=30):
    indexer = pd.api.indexers.FixedForwardWindowIndexer(window_size=dias)
    return df_grupo.shift(-1).rolling(window=indexer, min_periods=min_periods).sum()
```

```python
df['Venda_60_T'] = calcular_vendas_futuras_2(df.valor_venda)
```

```python
#Making target for 90 days
def calcular_vendas_futuras_3(df_grupo, dias=90, min_periods=30):
    indexer = pd.api.indexers.FixedForwardWindowIndexer(window_size=dias)
    return df_grupo.shift(-1).rolling(window=indexer, min_periods=min_periods).sum()
```

```python
df['Venda_90_T'] = calcular_vendas_futuras_3(df.valor_venda)
```

```python
df= df.dropna()
```

### Modeling 

#### 30 days

```python
df_30 = df.copy()
```

```python
df_30.columns
```

```python
df_30
```

```python
df_30.drop(columns = ['Venda_60_T', 'Venda_90_T'], inplace = True)
```

```python
df_30.info()
```

```python
subset = ['id_produto',
 'id_cupom',
 'id_cliente',
 'id_endereco_venda',
 'id_tipo_cliente',
 'cod_canal',
 'canal',
 'unidade_negocio',
 'cod_loja',
 'regional',
 'distrito',
 'cidade',
 'uf']
```

```python
#Enconding object class
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in subset:
  le.fit(df_30[i])
  df_30[i] = le.transform(df_30[i])

```

```python
#Setting the variables 
x = df_30.drop(columns= ['Venda_30_T','Data'])
y = df_30['Venda_30_T']
```

```python
# Making de split andTraining de model 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)

lr = LinearRegression()

lr.fit(x_train, y_train)
```

```python
r_sq = lr.score(x,y)
print(r_sq)
```

```python
y_pred_train = lr.predict(x_train)
print("MAE:", metrics.mean_absolute_error(y_train, y_pred_train))
```

```python
y_pred_test = lr.predict(x_test)
print("MAE:", metrics.mean_absolute_error(y_test, y_pred_test))
```

```python
y_pred = lr.predict(x)
```
